```python
import pandas as pd
import matplotlib.pyplot as plt
titanic_data = pd.read_csv('/content/train.csv')
titanic_data.head(10)
```

```python
ages = titanic_data['Age'].dropna()
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=30, edgecolor='k', alpha=0.7)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
survived_counts = titanic_data['Survived'].value_counts()
plt.figure(figsize=(8, 6))
survived_counts.plot(kind='bar', color=['red', 'green'], alpha=0.7)
plt.title('Survival Status of Titanic Passengers')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
pclass_counts = titanic_data['Pclass'].value_counts()
plt.figure(figsize=(8, 6))
pclass_counts.plot(kind='bar', color=['blue', 'orange', 'green'], alpha=0.7)
plt.title('Passenger Class Distribution on the Titanic')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
embarked_counts = titanic_data['Embarked'].value_counts()
plt.figure(figsize=(8, 6))
embarked_counts.plot(kind='bar', color=['purple', 'cyan', 'magenta'], alpha=0.7)
plt.title('Embarkation Ports of Titanic Passengers')
plt.xlabel('Embarkation Port')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
